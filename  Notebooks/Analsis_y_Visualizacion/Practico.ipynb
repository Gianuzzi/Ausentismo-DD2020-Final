# **Primera Aproximación a los datos**

## **Descripción del dataset**

El dataset forma parte de las bases de datos de la institución médica. Contiene información referida a los turnos correspondientes a consultas y prácticas médicas que se realizan en cada uno de los centros de atención.

### **Campos:**

##### **Información relativa al paciente**

* **Paciente:** Nombre y apellido del paciente (anonimizado)

* **Edad:** Edad del paciente

* **Sexo:** Sexo del paciente

* **Prestador:** Obra social, prepaga o consulta particular 

##### **Información relativa al turno**

* **CentroDeAtencion:** Cada uno de los cuatro centros de atención con que cuenta la institución

* **Servicio:** Especialidad médica correspondiente

* **Recurso:** Profesional o instrumento específico correspondiente al turno

* **PrestacionAsignada:** Prestación asignada en en el turno

* **EstadoDelTurno:** Estado (presente o pasado) correspondiente al turno. Indica si el turno fue atendido, se encuentra libre, si el paciente se ausentó, si fue cancelado o reprogramado, etc.

* **NombreDelCanal:** Canal de comunicación por donde se realizó la reserva del turno

* **FechaDelTurno:** Día correspondiente al turno

* **HoraDelTurno:** Horario del turno

* **EsSobreTurno:** Indica si se trata de un turno regular o de un sobre turno

* **TipoDeTurnoAsignado:** Indica si el turno corresponde a una primera consulta, a la realización de una práctica, a un control, etc.

```python
def get_null_fraction(df, column):
    return df[column].isnull().sum() / float(df.shape[0])

def get_null_percent(df, column):
    return 'NULL = {:.2f}%'.format(100*get_null_fraction(df, column))

def UnoSobreX(x, a):
    return a/x

def Lineal(x, a, b):
    return a*x + b

def Expon(x, a, b):
    return (x**a)*np.exp(b)

def OddsRatio(df, col1, a, col2, b):
    '''Calcula el Odds Ratio entre 2 variables de 2 columnas de un DataFrame.
    Devuelve: Odds Ratio, Desviación de Odds Ratio.'''
    if (type(a) == float) and np.isnan(a):
        p_a   = df[col1].isnull()
    else:
        p_a   = df[col1] == a
    if (type(b) == float) and np.isnan(b):
        p_b   = df[col2].isnull()
    else:
        p_b   = df[col2] == b
    P_a_b   = (p_a & p_b).sum()
    P_ac_b  = (~p_a & p_b).sum()
    P_a_bc  = (p_a & ~p_b).sum()
    P_ac_bc = (~p_a & ~p_b).sum()
    stdod   = np.sqrt(1/float(p_a.sum()) + 1/float((~p_a).sum()) + 1/float(p_b.sum()) + 1/float((~p_b).sum()))
    if (P_ac_b*P_a_bc == 0):
        return np.inf, stdod
    return (P_a_b*P_ac_bc)/float(P_ac_b*P_a_bc), stdod

def Bayes(df, col1, a, col2, b):
    '''Calcula la Probabilidad Condicional entre 2 variables de 2 columnas de un DataFrame.
    Devuelve: Probabilidad Condicional'''
    if (type(a) == float) and np.isnan(a):
        p_a   = (df[col1].isnull())
    else:
        p_a   = (df[col1] == a)
    if (type(b) == float) and np.isnan(b):
        p_b   = (df[col2].isnull())
    else:
        p_b   = (df[col2] == b)
    P_a_b = (p_a & p_b).sum()
    return P_a_b/float(p_b.sum())

def Bayes_Odds(df, col1, a, col2, b):
    '''Calcula la Probabilidad Condicional, y el Odds Ratio entre 2 variables de 2 columnas de un DataFrame.
    Devuelve: Probabilidad Condicional, Odds Ratio, Desviación de Odds Ratio.'''
    if (type(a) == float) and np.isnan(a):
        p_a   = (df[col1].isnull())
    else:
        p_a   = (df[col1] == a)
    if (type(b) == float) and np.isnan(b):
        p_b   = (df[col2].isnull())
    else:
        p_b   = (df[col2] == b)
    P_a_b   = (p_a & p_b).sum()
    P_ac_b  = (~p_a & p_b).sum()
    P_a_bc  = (p_a & ~p_b).sum()
    P_ac_bc = (~p_a & ~p_b).sum()
    bayes   = P_a_b/float(p_b.sum())
    stdod   = np.sqrt(1/float(p_a.sum()) + 1/float((~p_a).sum()) + 1/float(p_b.sum()) + 1/float((~p_b).sum()))
    if (P_ac_b*P_a_bc == 0):
        return  bayes, np.inf, stdod
    odds = (P_a_b*P_ac_bc)/float(P_ac_b*P_a_bc)
    return bayes, odds, stdod

def Zipfs(N, Nmax=0, s=1.07):
    '''Calcula las frecuencias asignadas por la distribución de Zipf.
    N puede ser un entero (cantidad de valores), o un array (índices).
    Nmax es el valor de cantidad asignado a la frecuencia inicial (máxima). Si Nmax=0, las frecuencias son normalizadas.
    Devuelve: Frecuencias'''
    if isinstance(N, int):
        N = np.arange(N) + 1
    f = 1/(N**s)/np.sum(1/(N**s))
    if Nmax!=0:
        return f*Nmax/float(f[0])
    return f

def kstest(observada, teorica, alpha=0.05):
    '''Calcula un test de Kolmogorov Smirnov entre 2 arrays.
    Ambos arrays deben tener la misma acumulada (ambos normalizado, o ninguno).
    La Hipótesis Nula (H0) es que ambas distribuciones son iguales, con el alpha impuesto.
    Devuelve: Ks Observado, Ks Teorico, pvalue'''
    if(np.abs(sum(observada)-sum(teorica))>1):
        raise Exception('Acumuladas distintas')
        return
    KsObs, pvalue = stats.ks_2samp(observada, teorica)
    KsTeo         = stats.kstwo.ppf(1-alpha, len(teorica)-1)
    Test          = KsObs < KsTeo
    print ('KS Observado = %.4f' %KsObs)
    print ('KS Teórico   = %.4f' %KsTeo)
    print ('%3.4f < %3.4f =>' %(KsObs, KsTeo), Test)
    if ~np.isnan(pvalue):
        print('p-value = %.2e' %pvalue)
    Test = alpha < pvalue
    if (Test):
        print ('Para alpha=%.2f, \n ¡No se puede desacreditar la hipótesis nula!'  %alpha)
    else:
        print ('Para alpha=%.2f, \n ¡Se rechaza la hipótesis nula!'  %alpha)
    return KsObs, KsTeo, pvalue

def chicuadrado(observada, teorica, alpha=0.05, normed=True): #BinWidth = 1
    '''Calcula un test de Kolmogorov Smirnov entre 2 arrays.
    Ambos arrays deben tener la misma acumulada (ambos normalizado, o ninguno).
    La Hipótesis Nula (H0) es que ambos array provienen de una misma distribución, con el alpha impuesto.
    Devuelve: Ks Observado, Ks Teorico, pvalue'''
    if(np.abs(sum(observada)-sum(teorica))>1):
        raise Exception('Acumuladas distintas')
        return
    ChiObs, pvalue = stats.chisquare(observada, teorica)
    ChiTeo         = stats.chi2.ppf(1-alpha, len(teorica)-1)
    Test           = ChiObs < ChiTeo
    print ('Chi Cuadrado Observado = %.4f' %ChiObs)
    print ('Chi Cuadrado Teórico   = %.4f' %ChiTeo)
    print ('%3.4f < %3.4f =>' %(ChiObs, ChiTeo), Test)
    if ~np.isnan(pvalue):
        print('p-value = %.2e' %pvalue)
    Test = alpha < pvalue
    if (Test):
        print ('Para alpha=%.2f, \n ¡No se puede desacreditar la hipótesis nula!'  %alpha)
    else:
        print ('Para alpha=%.2f, \n ¡Se rechaza la hipótesis nula!'  %alpha)
    return ChiObs, ChiTeo, pvalue
```

## Probabilidad condicional y Odds Ratio

### *Coeficientes*

* Probabilidad condicional
$$P(A|B) = \frac{P(A \cap B)}{P(B)} \equiv \frac{|A \cap B|}{|B|}$$

* Odds Ratio
$$OR(A,B) = \frac{|A \cap B|\times |A^{\mathsf{c}} \cap B^{\mathsf{c}}|}{|A^{\mathsf{c}} \cap B|\times |A \cap B^{\mathsf{c}}|}$$

$$STD(OR) = \sqrt{|A|^{-1} + |A^{\mathsf{c}}|^{-1} + |B|^{-1} + |B^{\mathsf{c}}|^{-1}}$$

```python
ProbCols = ['EstadoDelTurno', 'CentroDeAtencion', 'Sexo', 'Prestador', 'NombreDelCanal', 'EsSobreTurno', 'TipoDeTurnoAsignado',
             'Servicio', 'PrestacionAsignada']
ProbCols2 = ['EstadoDelTurno', 'CentroDeAtencion', 'Sexo', 'Prestador', 'NombreDelCanal', 'EsSobreTurno', 'TipoDeTurnoAsignado']
```

```python
OR    = []
St_OR = []
Bay   = []
lista = []
lcol  = []
lb    = []

col1  = 'EstadoDelTurno'
a     = 'Ausente'
for j in range(1, len(ProbCols)):
    col2 = ProbCols[j]
    print('%2.2f %% | %s' %((j-1)*100/float(len(ProbCols)-1), col2))
    for b in dfnn[Ausentesnn][col2].unique():
        lcol.append(col2)
        lb.append(b)
        lista.append(str(col2)[:7]+'_'+str(b))
        BAY, ODR, STO = Bayes_Odds(dfnn, col1, a, col2, b)
        Bay.append(BAY)
        OR.append(ODR)
        St_OR.append(STO)
#        print('(%s = %s | %s = %s)'
#              '\n Bayes  = %.4f'
#              '\n OR     = %.4f'
#              '\n St(OR) = %.4f'
#              %(col1, a, col2, b, BAY, ODR, STO))
print('%3.1f %%' %100)
OR    = np.asarray(OR)
St_OR = np.asarray(St_OR)
Bay   = np.asarray(Bay)
lista = np.asarray(lista)
lcol  = np.asarray(lcol)
lb    = np.asarray(lb)
```

```python
print('Diferencias entre Perfectos ([OR = inf] y [Bayes = 1]):', ((Bay == 1) == ~(OR  == np.inf)).sum())
Perfect = Bay == 1
Baye    = Bay[~Perfect]
OdR     = OR[~Perfect]
listar  = lista[~Perfect]
lcolr   = lcol[~Perfect]
lbr     = lb[~Perfect]
```

**Ajuste Lineal**

```python
lin_LogB_LogO = stats.linregress(np.log(Baye), np.log(OdR))
print(lin_LogB_LogO)
```

```python
plt.figure(dpi=130)
plt.title('Prob Condicional Vs. Odds Ratio')
plt.plot(Baye,OdR,'.k', label='Real')
plt.plot(Baye, Expon(Baye,lin_LogB_LogO[0],lin_LogB_LogO[1]),'-', color=RED, 
         label='$%.2fx + %.2f$ \n $r^2=%.3f$' %(lin_LogB_LogO[0], lin_LogB_LogO[1], lin_LogB_LogO[2]**2))
plt.xlabel('Probabilidad Condicional')
plt.ylabel('Odds Ratio')
plt.xscale('log')
plt.yscale('log')
plt.legend()
plt.show()
```

**Relación:**   
$$a\times\log\left(P(A|B)\right) + b = \log\left(OR(A,B)\right)$$

$$\equiv$$

$$P(A|B)^a\times e^b = OR(A,B)$$

# ZIPF'S LAW

Sea 
* $k$ el Ranking,
* $N$ el número de elementos,
* $s$ el valor de exponente caracterizando la distribución. $(s\simeq1.07);$

$f(k;s,N)=\dfrac{\dfrac{1}{k^s}}{\sum\limits^N_{n=1}\dfrac{1}{n^s}}$



Para todos los casos, intentaremos ajustar la curva de *Zipf* y una curva $a/x$ (a modo de aproximación).

Realizamos Tests de Hipótesis (Test de *Kolmogorov-Smirnov*), para comprobar si la distribiciónes observadas son acorde a la Ley de Zipf *(s=1.07)*.

Se trabajará con las PDF normalizadas.

Se calcula $\text{ks}_{obs}$, y en caso de que $\text{ks}_{obs}\gt \text{ks}_{(1-\alpha,\nu)} \equiv \alpha \gt$ *p_value* $\quad\Rightarrow$ Se rechaza $H_0$.

En todos los casos tendremos **Hipótesis nula $H_0$**: Ambas distribuciones son *iguales*, con $\alpha=0.05$.

## **Todos** (*Apariciones en dataset*)

```python
vecesXpaciente            = (dfnn.Paciente.value_counts(normalize=False)).value_counts(normalize=False)
vecesXpaciente            = pd.DataFrame(vecesXpaciente.rename("Cant_Pacie"))
vecesXpaciente.index.name = 'Cant_Veces'
vecesXpaciente            = vecesXpaciente.reset_index()
vecesXpaciente.index.name = 'Ranking'
```

```python
x = np.asarray(vecesXpaciente.index + 1)
y = np.asarray(vecesXpaciente.Cant_Pacie)
```

```python
print('Función a ajustar: a/x')
a_x, da_x = curve_fit(UnoSobreX, x, y/float(sum(y)))
print('a = %.3f +- %.3f   (%.2f%%)' %(a_x, np.sqrt(da_x), np.sqrt(da_x)*100/float(a_x)))
```

```python
yzip                         = Zipfs(x)
vecesXpaciente['~Zipf']      = (yzip*y[0]/float(yzip[0])).astype(int)
vecesXpaciente['Cant_Norm']  = y/float(sum(y))
vecesXpaciente['~Zipf_Norm'] = yzip
display(vecesXpaciente)
```

```python
plt.figure(figsize=(20,7))
plt.suptitle('Zipf\'s law', fontsize=20)

plt.subplot(1,2,1)
plt.title('PDF', fontsize=15)
plt.plot(x, y/float(sum(y)), '.--k', label='Observada')
plt.plot(x, yzip, '.-b', label='Zipf\'s Law')
plt.plot(x, UnoSobreX(x, a_x), '.--', color=RED, label='$%.2f/x$' %a_x)

plt.xscale('log')
plt.yscale('log')
plt.xlabel('Ranking de cantidad \n de pacientes')
plt.ylabel('Cantidad')
plt.legend(fontsize=12)

plt.subplot(1,2,2)
plt.title('CDF', fontsize=15)
plt.plot(x, np.cumsum(y)/float(sum(y)), '.--k', label='Observada')
plt.plot(x, np.cumsum(yzip), '.-b', label='Zipf\'s Law')

plt.xscale('log')
plt.xlabel('Ranking de cantidad \n de pacientes')
plt.ylabel('Acumulada')
plt.legend(fontsize=12)

plt.show()
```

### Test estadístico

```python
KsO, KsT, pvalue = kstest(y/float(sum(y)), Zipfs(x))
```
